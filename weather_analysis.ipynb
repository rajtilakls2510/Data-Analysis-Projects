# Analysis

```python
#1 Dependence of RISK_MM with RainToday and RainTomorrow
#2 Dependence of RainToday with RainTomorrow
#3 Dependence of Rainfall with RainTomorrow
#4 Dependence of Location with RainToday and RainTomorrow
#5 Dependence of WindGustDir and WindGustSpeed with RainTomorrow
#6 Dependence of WindDir9am and WindSpeed9am with RainTomorrow
#7 Dependence of WindDir3am and WindSpeed3am with RainTomorrow
#8 Dependence of Clouds with RainTomorrow
#9 Dependence of Evaporation with RainTomorrow
#10 Dependence of Sunshine with RainTomorrow
#11 Comparative dependences among highest dependences with RainTomorrow 
#12 Dependence of humidity with RainTomorrow
```

```python
#1 Dependence of RISK_MM with RainTomorrow
weather[(weather['RainTomorrow']=='NO')&(weather['RISK_MM']>=1.1)]
```

```python
weather[(weather['RainTomorrow']=='Yes')&(weather['RISK_MM']<=1.1)]
```

Looks like the minimum value for RISK_MM to cause a rain tomorrow should be 1.1. In other words, if the value of RISK_MM is greater than 1, it will surely rain tomorrow. But let's not limit ourselves to this prediction, but move on to do some other predictions 

```python
#2 Dependence of RainToday with RainTomorrow
weather['RainToday']=weather['RainToday'].astype('category',categories=["Yes",'No'])
weather['RainTomorrow']=weather['RainTomorrow'].astype('category',categories=['Yes','No'])
```

```python
weather['RainTodayCode']=weather['RainToday'].cat.codes+1
weather['RainTomorrowCode']=weather['RainTomorrow'].cat.codes+1
weather.head()
```

```python
weather.loc[:,['RainTodayCode','RainTomorrowCode']].corr()
```

The correlation value suggests that RainToday and RainTomorrow are not too much correlated. In simple words, it is not that if it rains today, it has to rain tomorrow or vice versa.

```python
# Which place receives highest rainfall?
weather.groupby('Location')['Rainfall'].agg(sum).sort_values(ascending=False).plot(kind='bar',figsize=(20,10))
plt.title('Top 20 places which receive highest rainfall')
plt.ylabel('Rainfall (mm)')
plt.show()
```

Cairns, Darwin and CoffsHarbour receive highest rainfall.  

```python
# Rainfall trend of Cairns
weather[weather['Location']=='Cairns'].plot(x='Date',y='Rainfall',figsize=(30,20))
plt.title("Rainfall trend of Cairns")
plt.ylabel("Rainfall (mm)")
plt.show()
```

It is the months of January through April when Cairns receives highest rainfalls. These rainfalls are so high in mm that it sometimes crosses the 200mm mark. Though amount of rainfall in Cairns has reduced significantly, it still receives its  highest rainfalls in some of the beginning months of each year.

```python
# Rainfall trends of Cairns in the year 2015
weather[(weather['Location']=='Cairns')&(weather['Date'].dt.year==2015)].plot(x='Date',y='Rainfall',figsize=(30,10))
plt.title("Rainfall trends of Cairns in the year 2015")
plt.ylabel("Rainfall (mm)")
plt.show()
```

On taking a close look at the year 2015, it can be seen that Cairns received highest rainfall during the month of February. It did receive considerably significant amount of rainfall during the months of December, January and March too. For the matter of the fact, it can be said that Cairns has its monsoon from December to March. 

```python
# Rainfall trend of Darwin
weather[weather['Location']=='Darwin'].plot(x='Date',y='Rainfall',figsize=(30,20))
plt.title("Rainfall trend of Darwin")
plt.ylabel("Rainfall (mm)")
plt.show()
```

It normally rains in Darwin from near the end of the year through the central months of the new year. One surprise entry to notice that Darwin has received heavy rains in the year 2011. Darwin's maximum rainfall data does not show that Darwin deserves to be in 2nd place. It is because of this surprise heavy rain in 2011 that Darwin has achieved the position. 

```python
# Darwin's rainfall trend in 2015
weather[(weather['Location']=='Darwin')&(weather['Date'].dt.year==2015)].plot(x='Date',y='Rainfall',figsize=(30,10))
plt.title("Darwin's rainfall trend in 2015")
plt.ylabel("Rainfall (mm)")
plt.show()
```

It can be analysed that Darwin has its maximum rainfall in the months of December and January. The monsoon generally ends with great showers in March.  

```python
#3 Dependence of Rainfall with RainTomorrow
weather.loc[:,['Rainfall','RainTomorrowCode']].corr()
```

The correlation value between Rainfall and RainTomorrow suggests that they are not correlated at all. In simple words, it is not that if it has rained a certain amount today, it will rain tomorrow.

```python
# Dependence of wind direction at 9 am with RainTomorrow at Cairns
weather[(weather['Location']=='Cairns')&(weather['RainTomorrow']=='Yes')].groupby('WindDir9am')['RainTomorrow'].agg('count').sort_values(ascending=False)
```

Whenever wind blows from S, SE, SSE and ESE, it rains tomorrow.

```python
# Dependence of wind direction at 3 pm with RainTomorrow at Cairns
weather[(weather['Location']=='Cairns')&(weather['RainTomorrow']=='Yes')].groupby('WindDir3pm')['RainTomorrow'].agg('count').sort_values(ascending=False)
```

Whenever wind blows from S, SE, SSE, ESE, it rains tomorrow.

For the matter of the fact, it can be predicted on which coast of Australia does Cairns lie. It is normally seen that Sea Breeze brings rainfall. On analyzing the direction of rain bringing wind, we can say that on which side of Cairns does the sea lie. From this, we will be able to say on which coast does Cairns lie.
Normally, Cairns' rain bringing winds blow from S, SE, SSE and ESE. It can be said that the sea to Cairns lie to the South of the City. Thus, it can be said that Cairns lie on the South coast.  

```python
# Dependence of wind gust speed with RainToday at Cairns
weather[(weather['RainToday']=='Yes')&(weather['Location']=='Cairns')].plot(x='Date',y='WindGustSpeed',figsize=(30,10))
plt.title("WindGustSpeed vs RainToday at Cairns")
plt.ylabel("Gust Speed")
plt.show()
```

It normally rains at Cairns when the wind gust speed exceed 20-30. Wind gust speed does not help us in determining whether it will rain tomorrow or not. It might happen that there are heavy gusts but no rain. This graph shows us that whenever it rains with gusty winds, how fast can the wind get. In simple words, how dangerous the winds become when it rains at Cairns. It can be seen that the winds can get upto 60. Once the winds got upto 80. 

```python
#8 Dependence of clouds with RainTomorrow at Cairns
weather[(weather['Location']=='Cairns')].loc[:,['Cloud9am','Cloud3pm','RainTomorrowCode']].corr()
```

The correlation values show that clouds at 9 am and clouds at 3pm are quite correlated but none of the cloud covers are correlated to Rain Tomorrow. In simple words, that if it is cloudy at 9am, it might remain cloudy at 3pm which has a rational explanation. But it does not mean that if there are clouds at 3pm, it will rain tomorrow.

```python
#9 Dependence of evaporation with RainToday and RainTomorrow at Cairns
weather[weather['Location']=='Cairns'].loc[:,['Evaporation','RainTodayCode','RainTomorrowCode']].corr()
```

Evaporation does not seem to have a significant relation with Rain today and Rain tomorrow. But the numbers suggest that evaporation has more effects on the condition of rain today rather than rain tomorrow. 

```python
#10 Dependence of sunshine with RainToday and RainTomorrow at Cairns
weather[weather['Location']=='Cairns'].loc[:,['Sunshine','RainTodayCode','RainTomorrowCode']].corr()
```

It is very very surprising to note that Rain today and tomorrow depends significantly on the values of sunshine. It is least probable that if there's sunshine there will be any rain today. But surprisingly, the correlation value denotes that the more is the sunshine, the more is the probability of rain. It might be because of the fact that more sunshine results in more humidity which ultimately ends up in transferring large amounts of water vapours to the atmosphere which later comes down as rain.

```python
#12 Dependence of humidity on RainToday and Tomorrow at Cairns
weather[weather['Location']=='Cairns'].loc[:,['Humidity9am','Humidity3pm','RainTodayCode','RainTomorrowCode']].corr()
```

The numbers suggests that humidity at 9am depends on humidity at 3pm. It is quite reasonable that the humidity does not fall quite rapidly such that if its 90% at 9am, it will fall to 20% at 3pm. If it goes down, it comes down quite slowly resulting in high amount of correlation value. But Humidity, either at 9am or 3pm, does not depend on Rain Today or Tomorrow. But according to this correlation, we get that Rain Today does depend on Rain Tomorrow to a certain extent.

# Trends

```python
# Humidity trends at Cairns in the year 2016
weather[(weather['Location']=='Cairns')&(weather['Date'].dt.year==2016)].plot(x='Date',y=['Humidity9am','Humidity3pm'],figsize=(30,10))
plt.title("Humidity trends at Cairns in the year 2016")
plt.ylabel("Humidity (%)")
plt.show()
```

Cairns is a coastal city that is the reason why it has an average humidity of almost 80%. The average humidity of Cairns rises as the monsoon season ends. There is not much change in the humidity throughout the year.

```python
# Humidity trend of Sydney in the year 2016
weather[(weather['Location']=='Sydney')&(weather['Date'].dt.year==2016)].plot(x='Date',y=['Humidity9am','Humidity3pm'],figsize=(30,10))
plt.title("Humidity trend of Sydney in the year 2016")
plt.ylabel("Humidity (%)")
plt.show()
```

```python
# Pressure trends at Melbourne in the year 2016
weather[(weather['Location']=='Melbourne')&(weather['Date'].dt.year==2016)].plot(x='Date',y=['Pressure9am','Pressure3pm'],figsize=(30,10))
plt.title("Pressure trends at Melbourne in the year 2016")
plt.ylabel("Pressure (Pa)")
plt.show()
```

```python
# Temperature trend at Perth in the year 2016
weather[(weather['Location']=='Perth')&(weather['Date'].dt.year==2016)].plot(x="Date",y=["MinTemp","MaxTemp"],figsize=(30,10))
plt.title("Temperature trend at Perth in 2016")
plt.ylabel("Temp (degree C)")
plt.show()
```

Australia experiences summer in the months from Nov through April. It experiences winter from May through Oct. At Perth, in summer, the temperatures can go upto 42 degree Centigrate and in winters, the temperatures can go down to 0 degree Centrigate.
